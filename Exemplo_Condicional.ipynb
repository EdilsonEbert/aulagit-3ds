```python
nota = float(input("Digite sua nota: "))

if nota > 6.0:
   print("Você foi APROVADO!")

elif nota >= 4.0 and nota < 6.0:
   print("Você está de RECUPERAÇÃO!")

else:
   print("Você foi REPROVADO!")
```
